#1, Generate Real Dataset

```python
from numpy.random.mtrand import f
import pandas as pd
import numpy as np

np.random.seed(42)

n = 1000

df = pd.DataFrame({
    'user_id': np.arange(1, n+1),
    'visit_date': pd.date_range(start='2024-01-01', periods=n, freq='h'),
    'device': np.random.choice(['mobile', 'desktop'], n, p=[0.7, 0.3]),
    'added_to_cart': np.random.choice([0,1], n, p=[0.4, 0.6]),
})

df['purchased'] = df['added_to_cart'] * np.random.choice([0,1], n, p=[0.5, 0.5])
df['revenue'] = df['purchased'] * np.random.randint(100, 2000, n)

df.to_csv('ecommerce_data.csv', index=False)


df.head()
```

#2. Understand Data

```python
df.info()
df.describe()
df.isnull().sum()
```

#3. Clean Data

```python
df['visit_date'] = pd.to_datetime(df['visit_date'])
df = df.drop_duplicates()
```

#4. Funnel Metrics

```python
total_users = df['user_id'].nunique()
cart_users = df[df['added_to_cart'] == 1]['user_id'].nunique()
buyers = df[df['purchased'] == 1]['user_id'].nunique()

cart_rate = cart_users / total_users
purchase_rate = buyers / total_users

print(cart_rate, purchase_rate)
```

#5. Drop-off Analysis

```python
drop_off = cart_users - buyers
print(drop_off)
```

#6. Device-Level Insights

```python
df.groupby('device').agg({
    'added_to_cart':'mean',
    'purchased':'mean'
})
```
